## Modelo de Reinforcement Learning para optimizar el precio de la cesta de la compra

La base de datos utilizada se ha creado de forma manual a modo de ejemplo con una serie de productos de Alcampo

Las funciones a las que se llama en el notebook pueden ser consultadas en el documento py.

referencia: https://www.kaggle.com/osbornep/reinforcement-learning-for-meal-planning-in-python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

import time

from funciones_RL import productos
from funciones_RL import MCModelv3
```

```python
data = pd.read_csv("./datos_alcampo.csv", sep = ';', encoding = 'latin')
```

```python
data
```

```python
data.groupby('Producto')['Precio'].mean()
```

```python
tabla, rewards, dinero = productos(data)
```

```python
budget11 = dinero
alpha11 = 0.05
num_episodes11 = 5000
epsilon11 = 0.2
reward11 = rewards
```

```python
start_time = time.time()

Mdl11 = MCModelv3(data=tabla, alpha = alpha11, e = num_episodes11,epsilon = epsilon11, budget = budget11, reward = reward11)

print("--- %s seconds ---" % (time.time() - start_time))

print(Mdl11[1])
```

```python
productos_optimos = Mdl11[4]
productos_optimos
```

```python
productos_mas_baratos = Mdl11[6]
productos_mas_baratos
```

```python
precio_optimo = Mdl11[5]
precio_optimo
```

```python
precio_mas_barato = Mdl11[7]
precio_mas_barato
```
