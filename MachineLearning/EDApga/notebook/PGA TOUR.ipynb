## Breve descripción sobre ciertos aspectos del golf

Antes de pasar a describir las variables hay una serie de conceptos que debes detallarse a cerca del golf para todos aquellos que no conozcan este deporte para que puedan entender las variables que se detallan a continuación.


- Par del campo: El par del campo de golf es el número de golpes en el que en teoría debe realizarse. Lo normal es que los campos tengan un par de 72 o 71 golpes. Si se hacen menos golpes se ha 'ganado' al campo.
- Par de cada hoyo: Es lo mismo que lo anterior pero para cada hoy. Hay 3 tipos diferentes de hoyos en función de su longitud y dificultad. Pares 3, los más cortos. Pares 4, son más largos que los pares 3, además suelen ser los más complicados. Pares 5, son los hoyos más largos, pero para los profesionales debido a su potencia suelen ser hoyos de recuperación, en los cuales se busca hacer menos de 5 golpes. Si se hace el hoyo en los golpes establecidos se dice que se ha hecho el par.
- Tee: El Tee de salida es la zona de cada hoyo desde la que se realiza el primer golpe.
- Calle: Zona principal del hoyo donde la hierba es más baja y desde donde es más fácil golpear la bola. El objetivo desde el Tee de salida es dejar la bola en calle con el primer golpe para los pares 4 y 5.
- Green: Es la zona del campo donde está el hoyo en el que hay que embocar la bola. Es una zona del campo con la hierba más baja de todo el hoyo, donde mejor rueda la bola y se usa un palo diferente a los demás, el putt.

## Descripción de las variables

- Player name: Nombre del jugador
- Rounds: Rondas jugadas en el año
- Fairway Percentage: Efectividad en coger calle desde el tee de salida
- Year: temporada
- Avg Distance: Distancia media
- gir: (Greens in regulation) La efectividad a la hora de llegar a green en los golpes en que debería hacerse en función del par del hoyo. Par 3 en 1 golpe, Par 4 en 2 golpes, Par 5 en 3 golpes
- Average Putts: Media de putts
- Average Scrambling: Recuperaciones tras fallar el llegar a green en los golpes 'habituales' haciendo par o mejor en el hoyo
- Average Score: Media de golpes por ronda
- Points: Puntos ganados en la FedexCup. Estos puntos se ganan a lo largo de la temporada en función de la posición final en cada torneo
- Wins: Veces que se ha ganado un torneo en la temporada
- Top 10: Veces que el jugador ha finalizado en un torneo en el Top 10 durante la temporada
- Average SG Putts: Media de golpes ganados o perdidos al campo en los putts
- Average SG Total: SG Off-the-tee + SG approach-the-green + SG around-the-green + SG putting statistics combined. Es la combinación de los golpes ganados o perdidos al campo teniendo en cuenta la suma de todas las estadísticas de golpes ganados o perdidos al campo en distintos apartados del juego. Estas estadísticas se detallan a continuación, ya que son también variables del data set
- SG:OTT: (Strokes Gained: Off-the-Tee) Golpes ganados o perdidos al campo desde el tee de salida en hoyos de par 4 o 5
- SG:APR: (Strokes Gained: Aproach) Golpes ganados o perdidos en el aproach a green. No se incluyen los golpes de salida en los pares 4 y 5, pero sí los de los pares 3, ya que se llega de un golpe a green
- SG:ARG: (Strokes Gained: Around-the-Green): Golpes ganados o perdidos alrededor del green, a menos de 30 yardas de distancia, sin tener en cuenta los golpes dentro de green (putts)
- Money: Dinero ganado en premios

## Delete some variables

- Points: Se elemina debido a que cuantos más puntos se hagan durante la temporada implica que más dinero se gana, ya que está directamente relacionado el dinero obtenido en un torneo con la posición en la que se queda, que a su vez implica un mayor número de puntos.

- Average Score: Existe una relación muy directa entre la media de golpes realizados durante la temporada y la posición del jugador en los torneo y, por tanto, de los premios acumulados. Por ello, no se considera como una variable adecuada para poder predecir el dinero en premios ya que no es una estadística concreta del juego de un jugador.
- Average SG Total: Se elimina ya que es una variable que tiene en cuenta otras 4 variables del dataset, y se desea saber el efecto que tienen dichas estadísticas del juego sobre el dinero ganado de forma individual.

```python
del pga_model['Points']
del pga_model['Average Putts']
del pga_model['Average Score']
del pga_model['Average SG Total']
```

## Variable predictors and predicted

```python
# endógena

endogena = ['Money'] 

# exógenas

exogenas = list(set(list(pga_model.columns))-set(endogena)) 
```

## Split the data (Train and test)

```python
X = pga_model[exogenas].values
Y = pga_model[endogena].values

# realizo una division del dataframe en train y test en una proporcion 80:20

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=40)
print(X_train.shape); print(X_test.shape)
```

# Linear Model

```python
regr = linear_model.LinearRegression()

# train the model
regr.fit(X_train, Y_train)
# predict
pred_linear_regr = regr.predict(X_test)
print(np.sqrt(mean_squared_error(Y_test,pred_linear_regr))) # error in test split
error_linear_regr = np.sqrt(mean_squared_error(Y_test,pred_linear_regr))
print(r2_score(Y_test, pred_linear_regr))
```

El modelo tiene un R^2 de 0.8433, y un error de 0.39.

```python
# DataFrame of the predictor's coefficients
regr.fit(X, Y)
coef_lista_regr = list(itertools.chain(*regr.coef_))
regr_coefficients = pd.DataFrame(coef_lista_regr, exogenas, columns = ['Coeficientes'])
regr_coefficients.loc['Intercept'] = regr.intercept_
regr_coefficients
```

# Lasso

```python
# optim value of alpha with cross validation

lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000, normalize = True)
lassocv.fit(X_train, Y_train)
lassocv.alpha_

model_lasso = Lasso(alpha=lassocv.alpha_) # lasso function
model_lasso.fit(X_train, Y_train) # train fit
pred_lasso = model_lasso.predict(X_test) # predict
print(np.sqrt(mean_squared_error(Y_test,pred_lasso))) # error in test split
error_lasso = np.sqrt(mean_squared_error(Y_test,pred_lasso))
print(r2_score(Y_test, pred_lasso)) # square R
```

El modelo tiene un R^2 de 0.843 y un error de 0.39 en la predicción con Lasso.

```python
# DataFrame of the predictor's coefficients
model_lasso.fit(X, Y)
lasso_coefficients = pd.DataFrame(model_lasso.coef_, exogenas, columns = ['Coeficientes'])
lasso_coefficients.loc['Intercept'] = model_lasso.intercept_
lasso_coefficients
```

# Ridge

```python
# values of alpha

alphas = 10**np.linspace(10,-2,100)*0.5

# optim value of alpha for the ridge regression to obtain the less MSE

ridgecv = RidgeCV(alphas = alphas, scoring = 'neg_mean_squared_error', normalize = True)
ridgecv.fit(X_train, Y_train)
ridgecv.alpha_

RidgeRegression = Ridge(alpha=ridgecv.alpha_) # funcion ridge
RidgeRegression.fit(X_train, Y_train) # ajuste sobre el train
pred_test_rr = RidgeRegression.predict(X_test) # prediccion sobre el test
print(np.sqrt(mean_squared_error(Y_test,pred_test_rr)))  # error sobre el test
error_ridge = np.sqrt(mean_squared_error(Y_test,pred_test_rr))
print(r2_score(Y_test, pred_test_rr)) # R^2 sobre el test
```

El modelo tiene un R^2 de 0.84 y un error de 0.3913 en la predicción con Ridge.

```python
# Coefficients

RidgeRegression.fit(X, Y)
coef_lista = list(itertools.chain(*RidgeRegression.coef_))
ridge_coefficients = pd.DataFrame(coef_lista, exogenas, columns = ['Coeficientes'])
ridge_coefficients.loc['Intercept'] = RidgeRegression.intercept_
ridge_coefficients
```

```python
# we create a serie where we put the different model errors
serie_error = pd.Series([error_lasso, error_ridge, error_linear_regr],
                       index = ['error_lasso', 'error_ridge', 'error_linear_regr'])
serie_error
```

```python
# less error
serie_error.idxmin()
```

El menor error obtenido es el del modelo Lasso.
