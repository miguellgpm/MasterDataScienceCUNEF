```python
X_test = loan.drop('target',axis = 1).values
Y_test = loan['target'].values
```

## 2. DEFINING FUNCTIONS

```python
def conf_matrix(Y_test, Y_pred):
    cnf_matrix_tra = confusion_matrix(Y_test, Y_pred)
    conf_plot = sns.heatmap(cnf_matrix_tra, annot=True, cmap='Blues', fmt='g')
    bottom, top = conf_plot.get_ylim()
    conf_plot.set_ylim(bottom + 0.5, top - 0.5)
    conf_plot.set_title("Confusion matrix, without normalization")
    conf_plot.set_ylabel('True Label')
    conf_plot.set_xlabel('Predicted Label');
```

```python
def show_results(test, predict):
    conf_matrix(test, predict)
    print(classification_report(test,predict))
```

```python
def show_ROC(predict, xtest, ytest):
    metrics.plot_roc_curve(predict, xtest, ytest)
    plt.show()                     
    print("Score: ", predict.score(xtest, ytest))
```

```python
def load_models(filename):
    with open(filename, 'rb') as file:
        return(pickle.load(file))
```

## 3. RESULTS

### Random Forest with PCA

```python
RM_PCA = load_models('../src/Models/WithPCA/random_forest_regression')
```

```python
pred_RM_PCA = RM_PCA.predict(X_test)
conf_matrix(Y_test, pred_RM_PCA)
print(classification_report(Y_test, pred_RM_PCA))
```

```python
show_ROC(RM_PCA, X_test, Y_test)
```

### Random Forest with PCA and t-SNE

```python
RM_PCA_TSNE = load_models('../src/Models/WithTSNE/random_forest_regression_TSNE')
```

```python
pred_RM_PCA_TSNE = RM_PCA_TSNE.predict(X_test2)
conf_matrix(Y_test2, pred_RM_PCA_TSNE)
print(classification_report(Y_test2, pred_RM_PCA_TSNE))
```

We will draw the ROC Curve and calculate the area under the ROC Curve to show the results. It also has been compared the score in the prediction between the sample of the train and test parts.

```python
show_ROC(RM_PCA_TSNE, X_test2, Y_test2)
```

### XGBoost with PCA and bayesian optimization

```python
XGB_PCA_BO = load_models('../src/Models/WithBO/xgboost_regression')
```

After that, and once the Random Forest model has been estimated, we predict and show the results through the confussion matrix.

```python
pred_XGB_PCA_BO = XGB_PCA_BO.predict(X_test)
conf_matrix(Y_test, pred_XGB_PCA_BO)
print(classification_report(Y_test,pred_XGB_PCA_BO))
```

Finally, we show the ROC curve and the area under the curve. Also, it has been compared the score in the prediction between the sample of the train and test parts.

# Conclusions



Here we have made a comparison between the 3 best models obtained with new types of analyses to reduce de dimension of the number of numerical variables and with the method of Bayesian Optimization to find the best hyperparameters.

The best 3 models were the following ones:

1. Random Forest with PCA
2. Random Forest with PCA and t-SNE
3. XGBoost with PCA and Byesian Optimization

The results between those models were so differents. The worst model was the XGBoost with Bayesian Optimization, that had an accuracy of 0.78, that is better than the Random Forest with PCA and t-SNE, but it had the worst f1-score (0.15) and in the AUC (0.67).

The second one is the Random Forest with PCA and t-SNE, with an accuracy of 0.76 and a f1-score of 0.54. In the f1-score, this model is the best of all of them.

The best model of all that have been tested is the Random Forest with PCA, in which we have obtained an accuracy of 0.84 and a f1-score of 0.43. This one has had a worse f1-score result than the previous one, but if we combained the metrics of accuracy, f1-score and the AUC (0.80), we thought that this is the model that we could use to predict the default cases.

The bayesian optimization is faster than the search grid optimization, but once we have tested it in the different models, the search grid has showed better results than the first one. The documentation does not explain clealry why the library runs fine, without any problem with some models but with other it completely fails. We conclude about bayesian optmization that is to early to implement this method over a serious work, but it can have a good future and if the errors are solved and the documentation is improved too.
